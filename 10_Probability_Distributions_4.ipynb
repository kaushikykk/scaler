```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
survey_1 = [35, 36, 33, 37, 34, 35]
np.mean(survey_1)
```

```python
survey_2 = [20, 37, 17, 50, 53, 33]
np.mean(survey_2)
```

```python
n = 6
bootstrapped_samples = np.random.choice(survey_1, size=n)
bootstrapped_samples
```

```python
np.mean(bootstrapped_samples)
```

```python
n = 6
bootstrapped_samples_2 = np.random.choice(survey_2, size=n)
bootstrapped_samples_2
```

```python
np.mean(bootstrapped_samples_2)
```

```python
bootstrapped_mean_survey_1 = []
for i in range(10000):
    bootstrapped_samples = np.random.choice(survey_1, size=n)
    bootstrapped_mean_survey_1.append(np.mean(bootstrapped_samples))
bootstrapped_mean_survey_1
```

```python
sns.histplot(bootstrapped_mean_survey_1)
```

```python
bootstrapped_mean_survey_2 = []
for i in range(10000):
    bootstrapped_samples_2 = np.random.choice(survey_2, size=n)
    bootstrapped_mean_survey_2.append(np.mean(bootstrapped_samples_2))
```

```python
sns.histplot(bootstrapped_mean_survey_2)
```

```python
# 95% CI

x1 = np.percentile(bootstrapped_mean_survey_1, 2.5)
x2 = np.percentile(bootstrapped_mean_survey_1, 97.5)
print(x1, x2)
```

```python
x1 = np.percentile(bootstrapped_mean_survey_2, 2.5)
x2 = np.percentile(bootstrapped_mean_survey_2, 97.5)
print(x1, x2)
```
